```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import pandas as pd

path = "F:/software/selenium/geckodriver.exe"
driver = webdriver.Firefox(executable_path=path)

driver.get("https://www.adamchoi.co.uk/overs/detailed")

all_match = driver.find_element(by="xpath", value='//label[@analytics-event="All matches"]')
all_match.click()

matches = driver.find_elements(by="tag name", value='tr')

date = []
home_team = []
score = []
away_team = []

for i in matches:
    date.append(i.find_elements(by="tag name", value="td")[0].text)
    home_team.append(i.find_elements(by="tag name", value="td")[1].text)
    score.append(i.find_elements(by="tag name", value="td")[2].text)
    away_team.append(i.find_elements(by="tag name", value="td")[3].text)


d = {'date':date,'home team':home_team ,'score':score, 'away team':away_team}
df = pd.DataFrame(d)
df.to_csv('data.csv', index=False)

#for scraping japan data

dropdown = Select(driver.find_element(by="id", value="country"))
dropdown.select_by_visible_text("Japan")


matches2 = driver.find_elements(by="tag name", value='tr')
date2 = []
home_team2 = []
score2 = []
away_team2 = []

for i in matches2:
    date2.append(i.find_elements(by="tag name", value="td")[0].text)
    home_team2.append(i.find_elements(by="tag name", value="td")[1].text)
    score2.append(i.find_elements(by="tag name", value="td")[2].text)
    away_team2.append(i.find_elements(by="tag name", value="td")[3].text)


d2 = {'date':date2,'home team':home_team2 ,'score':score2, 'away team':away_team2}
df2 = pd.DataFrame(d2)
df2.to_csv('japan_data.csv', index=False)




#driver.quit()
```
